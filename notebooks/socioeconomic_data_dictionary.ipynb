
## Datasets:

#### IPEDS (2022-23)
- **HD2022** – *Directory Information*
  - Institution control (public/private), sector, Carnegie classification, state, urbanicity, zip code
- **IC2022_AY** – *Student Charges for Academic Year Programs*
  - In-state/out-of-state tuition, room & board, required fees, price estimates for full-time undergrads
- **SFA2122_P1** – *Student Financial Aid - Part A*
  - Number/percent Pell recipients, average federal loan, average institutional grant, number receiving aid
- **SFA2122_P2** – *Student Financial Aid - Part B*
  - Aid information by income and dependency (where available)
- **EF2022A** – *Fall Enrollment by Race/Ethnicity*
  - Enrollment by race (e.g., Asian, Black, White, Hispanic), total enrollment
- **ADM2022** – *Admissions Data (if available)*
  - Acceptance rates, yield rates, standardized test submission (for selectivity/preparedness)
- **C2022_A** – *Completions by Program (CIP)*  
   - Awards/degrees conferred by major (CIP code), award level, gender, and race/ethnicity; useful for linking field of study to debt outcomes and demographic breakdowns

- **C2022_B** – *Completions by Race/Ethnicity*  
   - Number of students completing degrees, disaggregated by race/ethnicity and gender; enables analysis of racial/ethnic composition of graduates across institutions

- **C2022_C** – *Completions by Age, Gender, Race/Ethnicity*  
   - Adds age group to completions data (e.g., under 18, 18–24, 25–39, 40+); helpful for understanding outcomes for nontraditional and adult learners


#### College Scorecard (2022–23)
- **Primary source for student debt and institutional outcomes**
  - Institution-level dataset providing student loan metrics, earnings, demographics, and selectivity
  - Used for dependent variable and key predictors

- Key variables:
  - `UNITID`: *Unique institution ID* (join key with IPEDS) — *Categorical*
  - `DEBT_MDN`: *Median federal student loan debt* for completers (primary dependent variable) — *Continuous*
  - `DEBT_N`: Number of borrowers used in `DEBT_MDN` estimate — *Continuous*
  - `MD_EARN_WNE_P10`: Median earnings 10 years after entry (for long-term outcomes) — *Continuous*
  - `FIRST_GEN`: Percent first-generation students — *Continuous*
  - `PELL_EVER`: Percent ever receiving Pell Grants — *Continuous*
  - `DEPENDENT`: Percent of students classified as dependent — *Continuous*
  - `COSTT4_A`: Total cost of attendance for academic year (includes tuition, fees, living costs) — *Continuous*
  - `NPT4_PUB`, `NPT4_PRIV`: Net price by income bracket (disaggregated by public/private) — *Continuous*
  - `SAT_AVG`: Average SAT score — *Continuous*
  - `ADM_RATE`: Admission rate — *Continuous*
  - `COMPL_RPY_3YR_RT`: 3-year repayment rate — *Continuous*
  - `UGDS`, `UGDS_BLACK`, `UGDS_WHITE`, `UGDS_HISP`, `UGDS_ASIAN`: Enrollment by race/ethnicity — *Continuous*
  - `CONTROL`, `REGION`, `LOCALE`: Institutional control, region, and geographic setting — *Categorical*

- These fields enable modeling of student debt against institutional, economic, and demographic factors




#### SHEF (FY2023 rows)
- **State Higher Education Finance dataset (FY2023, published 2024)**
  - Provides detailed state-level context on public investment in higher education
  - Used to evaluate how state funding levels, aid, and economic adjustments influence institutional costs and student borrowing outcomes

- Key variables (all from "Report Data" sheet):
  - `State`: U.S. state name — *Categorical*
  - `FY`: Fiscal year of data reporting — *Categorical*
  - `Education Appropriations`: Direct state support for public institution operations — *Continuous*
  - `Net Tuition and Fee Revenue`: Tuition revenue net of institutional discounts and waivers — *Continuous*
  - `Total State Support`: Combined appropriations including special-purpose allocations — *Continuous*
  - `State Public Financial Aid`: Total state-funded student financial aid — *Continuous*
  - `State Public Financial Aid as a Percent of Education Appropriations`: Proxy for state affordability effort — *Continuous*
  - `COLI (Cost of Living) Adjustment`: Index to normalize geographic cost differences — *Continuous*
  - `EMI (Enrollment Mix) Adjustment`: Adjustment for differences in student mix (e.g., undergrad vs. grad) — *Continuous*
  - `HECA (Inflation) Adjustment`: Higher Education Cost Adjustment for inflation-normalized comparisons — *Continuous*
  - `Net FTE Enrollment`: Full-Time Equivalent student count used for per-student metrics — *Continuous*
  - `Student Share`: Proportion of education revenue covered by students via tuition — *Continuous*
